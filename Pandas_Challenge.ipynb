```python
df["Salary"] = df["Salary"].astype(str).str.replace(r"[^\d.]", "", regex=True)
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
```

```python
df.head(20)
```

```python
df.info()
```

```python
df.tail(20)
```

```python
df["Age"] = df["Age"].astype(str).str.replace("-","", regex=False)
```

```python
df[df["Age"].str.contains("[a-zA-Z]",regex=True, na=False)]["Age"].unique()
```

```python
df["Age"] = df["Age"].str.replace("yrs","", regex=False).str.replace("twenty five","25", regex=False)
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
```

```python
df.isnull().sum()
```

```python
import numpy as np
df.loc[(df["PerformanceRating"] < 1) | (df["PerformanceRating"] > 5), "PerformanceRating"]=np.nan
df["PerformanceRating"] = df["PerformanceRating"].fillna(df["PerformanceRating"].median())
```

```python
df["ExperienceYears"].unique()
```

```python
df["ExperienceYears"] = df["ExperienceYears"].astype(str).replace("five","5", regex=False)
df["ExperienceYears"] = pd.to_numeric(df["ExperienceYears"], errors="coerce")
df["ExperienceYears"] = df["ExperienceYears"].fillna(df["ExperienceYears"].mean())
df["ExperienceYears"] = df["ExperienceYears"].round(0).astype(int)
```

```python
df.isnull().sum()
```

```python
df.tail(20)
```

```python
df["City"].unique()
```

```python
df["City"] = df["City"].str.strip()
df["City"] = df["City"].str.title()
df["City"] = df["City"].fillna("Unknown")
```

```python
df["Email"] = df["Email"].fillna("Not Provided")
```

```python
df["JoinDate"] = pd.to_datetime(df["JoinDate"], format="mixed", dayfirst=False, errors="coerce")
```

```python
df.isnull().sum()
```

```python
df.info()
df.shape
```

```python
df["Age"] = df["Age"].round(0)
df["Salary"] = df["Salary"].round(2)
```

```python
df["Department"] = df["Department"].str.strip().str.title()
```

```python
df["Department"].unique()
```

```python
df["Department"] = df["Department"].replace({"It": "IT", "Hr": "HR"})
```

```python
df.to_excel("cleaned_employee_data.xlsx", index=False)
```
